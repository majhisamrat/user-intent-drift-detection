```python
import pandas as pd
```

```python
df = pd.read_csv("data/raw/online_retail_II.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["Country"].value_counts()
```

```python
df = df.dropna(subset=["Description", "Customer ID"])
df["Description"] = df["Description"].str.lower()
```

```python
df["total_value"] = df["Quantity"] * df["Price"]
```

```python
def assign_intent(row):
    if row["Quantity"] >= 30:
        return "bulk_purchase"
    elif row["total_value"] >= 500:
        return "high_value_purchase"
    elif "christmas" in row["Description"]:
        return "seasonal_purchase"
    elif row["Country"] != "United Kingdom":
        return "international_purchase"
    else:
        return "normal_purchase"

df["intent"] = df.apply(assign_intent, axis=1)
```

```python
intent_df = df[["Description", "intent"]].rename(
    columns={"Description": "text"}
)

intent_df["intent"].value_counts()
```
